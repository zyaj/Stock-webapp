```python
import requests
import pandas as pd
import datetime
import dateutil
import re
from bokeh.models import Range1d

from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.embed import components
```

```python
tickersymbol = 'AAPL'
url = 'https://www.quandl.com/api/v3/datasets/WIKI/' + tickersymbol + '.json'
r = requests.get(url)
```

```python
request_df = pd.DataFrame(r.json())
request_df
```

```python
df = request_df['dataset'].apply(pd.Series) 
df.head(5)
```

```python
data = df.ix['data',:].apply(pd.Series) 
data.head(5)
```

```python
data.columns = df.ix['column_names',0:12]
data.head(5)
```

```python
data['Date'] = pd.to_datetime(data['Date'])
data.dtypes
data.head(5)
```

```python
output_notebook()
```

```python

today = datetime.date.today()
today
```

```python
lastmonth = today - dateutil.relativedelta.relativedelta(months=1)
lastmonth
```

```python
b = data['Date']>lastmonth
b.head(25)
```

```python
data_lastmonth = data[b]
data_lastmonth
```

```python
p = figure(title=tickersymbol,plot_width=600,plot_height=400,x_axis_type="datetime",x_range=Range1d(lastmonth, today))
p.line(data_lastmonth['Date'],data_lastmonth['Open'])
p.line(data_lastmonth['Date'],data_lastmonth['Open'],line_color='blue',legend='Opening price')
p.xgrid.minor_grid_line_color = 'navy'
```

```python
show(p)
```

```python
script, div = components(p)
```

```python
import numpy as np
req = 'https://www.quandl.com/api/v3/datasets/WIKI/'
req = '%s%s.json?&collapse=weekly' % (req,"AAPL")
req = '%s&start_date=%s-01-01' % (req,2016)
r = requests.get(req)

# transfer json data to date frame
cols = r.json()['dataset']['column_names'][0:5]
df = pd.DataFrame(np.array(r.json()['dataset']['data'])[:,0:5],columns=cols)
df.Date = pd.to_datetime(df.Date)
df[['Open','High','Low','Close']] = df[['Open','High','Low','Close']].astype(float)
```
